```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt
```

```python

data = {'Feature': [3, 5, 7, 9,9.5,10], 'Target': [8, 12, 14, 18,18.5,18.6]}
df = pd.DataFrame(data)
```

```python
# Fit a Decision Tree Regressor
tree_regressor = DecisionTreeRegressor()
tree_regressor.fit(df[['Feature']], df['Target'])
plt.scatter(df['Feature'], df['Target'])
```

```python
# Plot the decision tree
plt.figure(figsize=(10, 6))
plot_tree(tree_regressor, feature_names=['Feature'], filled=True, rounded=True)
plt.title("Decision Tree for Regression Example")
plt.show()
```
